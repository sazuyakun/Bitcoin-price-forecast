# Making the final data

```python
import pandas as pd

btc_df = pd.read_csv("Bitcoin Historical Data.csv")
# epu_df = pd.read_csv("EPU.csv")
```

```python
# Filter for rows where Year is 2024 AND Month is 4.0
epu_df[(epu_df["Year"] == 2024) & (epu_df["Month"] == 4.0)]
```

```python
btc_df.head()
```

```python
btc_df["Month"] = btc_df["Date"].apply(lambda x: float(x[3:5]))
btc_df["Year"] = btc_df["Date"].apply(lambda x: int(x[6:10]))
```

```python
btc_df.head()
```

```python
# btc_df["EPU"] is a col that takes value from epu_df["GEPU_current"] by matching the btc_df["Year"] and btc_df["Month"] with epu_df["Year"] and epu_df["Month"]
# we use def function instead of lambda function
def get_epu(year, month):
    epu_row = epu_df[(epu_df["Year"] == year) & (epu_df["Month"] == month)]
    if not epu_row.empty:
        return epu_row["GEPU_current"].values[0]
    else:
        return None

btc_df["EPU"] = btc_df.apply(lambda row: get_epu(row["Year"], row["Month"]), axis=1)
```

```python
btc_df.drop(columns=["Month", "Year"], inplace=True)
```

```python
btc_df.to_csv("Btc_with_epu.csv", index=False)
```

# Gold pricing

```python
import yfinance as yf

gold = yf.download("GC=F", start="2014-04-01", end="2024-04-02")  # Gold Futures
# gold = gold[['Adj Close']].rename(columns={'Adj Close': 'Gold'})
gold.reset_index(inplace=True)
```

```python
gold
```

```python
# Create a complete date range
date_range = pd.date_range(start="2014-04-01", end="2024-04-01", freq='D')
gold_complete = pd.DataFrame({'Date': date_range})

# Flatten the MultiIndex columns
gold.columns = gold.columns.get_level_values(0) if gold.columns.nlevels > 1 else gold.columns

# Merge with gold data and forward fill missing values
gold_complete = gold_complete.merge(gold[['Date', 'Close']], on='Date', how='left')
gold_complete = gold_complete.rename(columns={'Adj Close': 'Gold'})
gold_complete['Close'] = gold_complete['Close'].fillna(method='ffill')  # Forward fill weekend/holiday gaps
```

```python
len(gold_complete)
```

```python
gold_complete = gold_complete[::-1].reset_index(drop=True)
```
